# Estruturas de repetição


Estrutura de repetição permitem executar um comando até que determinada condição seja atendida


```python
#Criando uma lista com os numeros de 1 a 20
lista = list(range(1,20))
```

```python
print(lista[0])
print(lista[1])
```

```python
#Percorrendo todos os elementos da lista e mostrando junto com uma mensagem
for i in lista:
  print("O valor de lista[{}] é:{}".format(i,lista[i-1]))
```

```python

#Somando 10 a todos os numeros da lista
for i in lista:
  print(i+10)
```

```python
print(lista)
```

```python
# o for permite que seja analisado item a item de uma lista
#Script para encontra o maior elemento da lista
maior=0
for i in lista:
  if i > maior:
    maior = i

print("O Maior valor é: ",maior)
```

```python
#Script para verificar o menor elemento da lista
menor=1000
for i in lista:
  if i < menor:
    menor = i
print(menor)
```

```python
#imprimindo todos os nomes armazenados na agenda
for i in agenda.keys():
  print("Nome:",i)
  print("Telefone:",agenda[i][0])
  print("Email:",agenda[i][1])
  print("====================")
```

Estruturas de repetição permitem iterar entre os diversos itens de uma lista, aplicando a estes operações de soma, subtração, divisão ou qualquer outra ação que o python permitir como remoçao de dados



```python
#Para iterar entre os objetos pode ser utilizado os comandos while e for
```

```python
a = 0
while a<10:
  print(a)
  a=a+1
```

```python
#soma dos numeros de 1 a 10

a=10
soma=0
while a>1:
  soma=soma+a
  a=a-1
print(soma)
```

```python
#Muito cuidado ao utilizar o while, pois é necessário gerar uma condição de parada
```

```python
#soma dos numeros de 1 a 10

a=10
soma=0
while a>1:
  soma=soma+a
  a=a-1
print(soma)
```

```python
soma=0
impares = [1,3,5,9,11]
for i in impares:
  soma=soma+i
print(soma)

```

```python
#tem um jeito fácil
soma=sum(impares)
print(soma)
```
